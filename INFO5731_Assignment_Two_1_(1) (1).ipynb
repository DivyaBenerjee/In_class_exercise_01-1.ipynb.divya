```python
import requests
import csv
from bs4 import BeautifulSoup


A =[]
B =[item*10 for item in range(1,11)]

for item in B:
  URL = 'https://citeseerx.ist.psu.edu/search?q=natural+language+processing&t=doc&sort=rlv&start={}'.format(item)
  page = requests.get(URL)
  soup = BeautifulSoup(page.content, 'html.parser')
  for item in soup.find_all("div",{"class":"snippet"}):
    A.append(item.text)
```

```python
import pandas as pd
df = pd.DataFrame (A, columns = ['Abstract'])
df.to_csv("HundreadAbstracts.csv")
df
print(df)
```

```python
import re
import nltk
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import Word

# Removing of punctuations and special characters
df['Noise_Removal'] = df['Abstract'].str.replace('[^\w\s]','')
df['Noise_Removal'] = df['Noise_Removal'].apply(lambda item: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', char) for char in item ))

# Removing of digits
df['Digit_Removal'] = df['Noise_Removal'].apply(lambda x: ''.join([item for item in x if not item.isdigit()]))

#  removal of stop words
stopwords = stopwords.words('english')
df['Stopwords_Removal'] = df['Digit_Removal'].apply(lambda item: " ".join(item for item in item.split() if item not in stopwords))

#  lower Case comversion
df['To_Lower_Case'] = df['Stopwords_Removal'].apply(lambda item: " ".join(item.lower() for item in item.split()))

df
```

```python
# Stemming
st = PorterStemmer()
df['Stemming'] = df['To_Lower_Case'].apply(lambda x: " ".join([st.stem(X) for X in item]))

# Lemmatization
df['Lemmatization'] = df['Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```
